```python
#integrated. step wise change. modfications based on each extremum points
# works better_Main

from pyfirmata import Arduino, SERVO
from time import sleep
import keyboard
import time
import numpy as np
import pandas as pd
import os
import cv2 as cv
from threading import Thread
import datetime
import winsound

def servofunction():
    
    global deltamax, deltamin, MAX, MIN, step, calculations, calfreq, step, sleep_time, emerexit, freqcalculations, emerexit2, emerexit3, emerexit4, emerexit5, wt, ct, finishtime, minanglisttotal, maxanglisttotal, frequencytotal, finish, finishind

    #port = 'COM5'
    #pin = 9
    board = Arduino('COM3')

    #board.digital[pin].mode = SERVO

    servo = board.get_pin('d:9:s')

    end = False
    avefeq = True
    freqcalculations = False
    
    
    deltamax = 0
    deltamin = 0
    calculations = False
    #sleep_time = float(input("sleep time (0.078 for 1 Hz frequency):"))
    #MIN = int(input("what is the minimum angle:"))
    minang = MIN + 90
    minanglist = np.array([])
    #MAX = int(input("what is the maximum angle:"))
    maxang = MAX + 90
    maxanglist = np.array([])
    
    CBLUE = '\033[34m'
    CGREEN = '\033[92m'
    CEND = '\033[0m'
    CRED = '\033[91m'
    
    t1totalcal = True
    t2totalcal = True
    
    alarm = True
    
    step = 5
    #step = int(input("what is the step angle:"))
    
    sleep_time = 0.078
    #sleep_time = float(input("sleep time (0.078 for 1 Hz frequency):"))

    def rotateServo(angle):
        global sleep_time
        servo.write(angle)
        sleep(sleep_time)

    print("                 Servo ready to start")
    while True:
        if keyboard.is_pressed("r"):
            print("Servo Starts")
            for i in range(90, minang, -step):
                    rotateServo(i)  

            #t = np.array([time.time()])
            if calfreq == True:
                frequency = np.array([])

            loopN = 1

            while True:
                
                ct1 = time.time()
                
                if emerexit == True:
                    rotateServo(90)
                    break
                
                if emerexit4 == True:
                    rotateServo(90)
                    break
                    
                if emerexit5 == True:
                    rotateServo(90)
                    break
                    
                if emerexit2 == True:
                    rotateServo(90)
                    break
                
                step = cv.getTrackbarPos("step", "Servo controller")
                if step == 0:
                    step = 1
                sleep_time = (cv.getTrackbarPos("sleep_time", "Servo controller"))/1000
                if sleep_time < 0.01:
                    sleep_time = 0.01


                if deltamax > 7:
                    maxang = maxang + 4
                elif deltamax > 5:
                    maxang = maxang + 3
                elif deltamax > 3:
                    maxang = maxang + 2
                elif deltamax > 1:
                    maxang = maxang + 1
                elif deltamax > -1:
                    pass
                elif deltamax > -3:
                    maxang = maxang - 1
                elif deltamax > -5:
                    maxang = maxang - 2
                elif deltamax > -7:
                    maxang = maxang - 3
                else:
                    maxang = maxang - 4
                if deltamin > 7:
                    minang = minang + 4
                elif deltamin > 5:
                    minang = minang + 3
                elif deltamin > 3:
                    minang = minang + 2
                elif deltamin > 1:
                    minang = minang + 1
                elif deltamin > -1:
                    pass
                elif deltamin > -3:
                    minang = minang - 1
                elif deltamin > -5:
                    minang = minang - 2
                elif deltamin > -7:
                    minang = minang - 3
                else:
                    minang = minang - 4

                #print("\n SERVO Loop #%d" % loopN,"   \n")
                if calculations == True:
                    print(CBLUE + "max ang on servo=", maxang, "   min ang on servo=", minang, "" + CEND)
                    maxanglist = np.append(maxanglist, maxang)
                    minanglist = np.append(minanglist, minang)

                #if maxang > 140:
                 #   emerexit3 = True
                    #print('servo is touching the device, max ang is set to 140')
                    #maxang = 140
                
                #if minang < 30:
                 #   emerexit3 = True
                    #print('servo is touching the device, min ang is set to 30')
                    #minang = 30
                    
                if emerexit3 == True:
                    rotateServo(90)
                    print("Servo touched the device")
                    break
                    
                if maxang == minang or maxang < minang:
                    print("ERROR")
                    break

                if calfreq == True:
                    t1=time.time()
                
                if t1totalcal == True:
                    t1total = time.time()
                    t1totalcal = False
                
                for i in range(minang, maxang, step):
                    rotateServo(i)

                    #print(count, "applied angels =", i, " actual angel =", servo.read(), "time =", t[count])

                    if keyboard.is_pressed("q"):
                        cp = servo.read()
                        if cp < 90:
                            #print(cp)
                            for j in range (cp, 91, step):
                                rotateServo(j)
                        elif cp > 90:
                            #print(cp)
                            for j in range (cp, 89, -step):
                                rotateServo(j)
                        else:
                            pass
                        #print("q is pressed 60 to 120")
                        end = True
                        break
                if end == True:
                        print("press ", CRED + "E" + CEND," to stop servo")
                        break
                for i in range(maxang, minang-1, -step):
                    rotateServo(i)
                    #print(count, "applied angels =", i, " actual angel =", servo.read(), "time =", t[count])

                    if keyboard.is_pressed("q"):
                        cp = servo.read()
                        if cp > 90:
                            #print(cp)
                            for j in range (cp, 89, -step):
                                rotateServo(j)
                        elif cp < 90:
                            #print(cp)
                            for j in range (cp, 91, step):
                                rotateServo(j)
                        else:
                            pass
                        #print("q is pressed 120 to 60")
                        end = True
                        break
                
                if calfreq == True:
                    t2=time.time()
                
                if end == True:
                        print("press ", CRED + "E" + CEND," to stop servo")
                        break
                
                ct2 = time.time()
                tcurrent = ct2 - ct1
                #currenttime = datetime.timedelta(seconds=tcurrent)
                ct = np.append(ct, datetime.timedelta(seconds=tcurrent))
                if ct.sum() > finishtime:
                    emerexit5 = True
                    finishind = True
                if ct.sum() > datetime.timedelta(seconds=780) and alarm == True:
                    alarm = False
                    winsound.MessageBeep()
                    
                
                
                print("The total time so far is ", str(ct.sum()), " (Hours:Minutes:Seconds)")
                
                
                #a = int(((((maxang-minang)*2)/10)+1))
                if calfreq == True and freqcalculations == True:
                    fre = 1/(t2-t1)
                    frequency = np.append(frequency, fre)
                    #print("t loop begin=", t[count-13]," t loop end=", t[count-1])
                    print(CGREEN + "Frequency= ", frequency[-1], "" + CEND)
                    loopN = loopN + 1

        if calfreq == True and calculations == True and avefeq == True and finishind == True:
            dffrequency = pd.DataFrame(frequencytotal)
            dffrequency.rename(columns={0: 'frequencies'}, inplace=True)
            dffrequency.to_excel("frequency.xlsx", float_format = "%.3f", header = True)
            #mF = (sum(frequencytotal)-1)/(frequencytotal.size-1)
            #print("\n the average Frequency = ", mF)
            
        if calfreq == True and calculations == True and avefeq == True and finishind == False:
            frequencytotal = np.append(frequencytotal, frequency)  
            
        if calculations == True and avefeq == True and finishind == True:
            print('servo data is saved')
            dfmaxanglist = pd.DataFrame(maxanglisttotal)
            dfmaxanglist.rename(columns={0: 'maximums on servo'}, inplace=True)
            dfmaxanglist.to_excel("max angs on servo.xlsx", float_format = "%.3f", header = True)
            dfminanglist = pd.DataFrame(minanglisttotal)
            dfminanglist.rename(columns={0: 'minimums on servo'}, inplace=True)
            dfminanglist.to_excel("min angs on servo.xlsx", float_format = "%.3f", header = True)
            avefeq = False
            
        if calculations == True and avefeq == True and finishind == False:
            maxanglisttotal = np.append(maxanglisttotal, maxanglist)
            minanglisttotal = np.append(minanglisttotal, minanglist)
        
        if t2totalcal == True and t1totalcal == False:
            t2total = time.time()
            t2totalcal = False
            ttotal = t2total - t1total
            wholetime = str(datetime.timedelta(seconds=ttotal))
            print("The cycle process time was ", wholetime, " (Hours:Minutes:Seconds)")
            #wt = np.append(wt, datetime.timedelta(seconds=ttotal))
            #print("The total time is ", str(wt.sum()), " (Hours:Minutes:Seconds)")
            print("The total time is ", str(ct.sum()), " (Hours:Minutes:Seconds)")
        
        if keyboard.is_pressed("e") or emerexit == True or emerexit2 == True or emerexit3 == True or emerexit4 == True or emerexit5 == True:
            #print("servo about to stop")
            board.exit()
            break
    print("servo stops")
    


                
                
                
    
def monitorfunction():
    
    global rect,startPoint,endPoint, assign, MAX, MIN, deltamax, deltamin, calculations, recvid, step, sleep_time, emerexit, freqcalculations, emerexit2, emerexit3, emerexit4, emerexit5, zoomed, finishind, finish, repeat, wt, ct, finishtime, hx, hy, hingselection, anglestotal, avemaxtotal, avemintotal, maxpeakstotal, minpeakstotal
    
    #clear the folder for the frames:
    #dir = 'live_frames'
    #for f in os.listdir(dir):
     #   os.remove(os.path.join(dir, f))
    angles = np.array([])
    avemax = np.array([])
    avemin = np.array([])
    maxpeaks = np.array([])
    minpeaks = np.array([])
    
    
    #live video:
    livevid = cv.VideoCapture(1+cv.CAP_DSHOW)
    
    livevid.set(cv.CAP_PROP_FRAME_WIDTH, 1920)
    livevid.set(cv.CAP_PROP_FRAME_HEIGHT, 1080)

    if recvid == True:
        fourcc = cv.VideoWriter_fourcc(*'XVID')
        savef = cv.VideoWriter('source.avi', fourcc, 30, (640,  480))
        
    def empty(a):
        pass
    
    def sething(event,x,y,flags,params):
        global rect, hx, hy, hingselection
        if event == cv.EVENT_LBUTTONDOWN:
                #rect = (x, y)
                if hingselection == True:
                    hx = x
                    hy = y
                    print("hinge point is selected: (", hx, ", ", hy, ")")
                    hingselection = False
                print("Now press S to continue:")

    cv.namedWindow("Servo controller")
    cv.resizeWindow("Servo controller", 640,80)
    cv.createTrackbar("step", "Servo controller", 3, 20, empty)
    cv.createTrackbar("sleep_time", "Servo controller", 20, 100, empty)
    
    tracker = cv.TrackerCSRT_create()
    
    #timeout = time.time() + 5
    
    if hingselection == True:
        print("Adjust the servo parameters, align the hing point with RED dot and click on the hing point (960, 468):")
    if hingselection == False:
        print("hing point is the same, press S to continue")
        

    
    while True:
        
        if keyboard.is_pressed("Esc"):
            emerexit5 = True
            finishind = True
            
        if emerexit5 == True:
            break
        
        liveret, livef = livevid.read()
        #livef = cv.medianBlur(livef,5)    
        #livehsv = cv.cvtColor(livef, cv.COLOR_BGR2HSV)

        cv.circle(livef, (960, 468), radius=1, color=(0,0,255), thickness=-1)
        
        cv.circle(livef, (960, 695), radius=2, color=(255,0,0), thickness=-1)
        cv.line(livef, (930,695), (990, 695), (125,125,0), 1, 1)
        cv.rectangle(livef, (910, 670), (1010, 720), (125,125,0), 1, 1)
        
        if zoomed == True:
            height, width, channels = livef.shape
        
            minY = 368
            maxY = 768
            minX = 604
            maxX = 1315
        
        
            cropped = livef[minY:maxY , minX:maxX]
            livef = cv.resize(cropped, (width, height))
        
        
        #cv.imshow('live hsv:', livehsv)
        cv.imshow("live video:", livef)
        
        if hingselection == True:
            cv.setMouseCallback("live video:", sething)
        
        if cv.waitKey(33) == ord('s'):
            break
        
    if emerexit5 == False:
        print("choose object for tracking")
        #cv.rectangle(livehsv, (hx-2, hy-2), (hx+2, hy+2), (255,0,255), 3, 1)
        #cv.line(livehsv, (hx, hy), (hx, 480), (0,255,0), 3, 1)
        #cv.line(livehsv, (hx, hy), (hx, 480), (255,0,255), 1, 1)
        #cv.line(livehsv, (hx-28, hy+350), (hx+28, hy+350), (255,0,255), 2, 1)
        #cv.rectangle(livef, (323, 4), (327, 8), (255,0,255), 3, 1)
        bbox = cv.selectROI("live video:", livef, False)
        print("Coordinates for the center of the tracking object are (", bbox[0]+(bbox[2]/2),", ", bbox[1]+(bbox[3]/2),")")
    
    
        aa = hx-(bbox[0]+(bbox[2]/2))
        if aa<10 and aa>-10:
            print("The hinge and tracking center are aligned")
            pass
        else:
            print("The hinge and tracking center are NOT aligned")
            emerexit2 = True
        
    
        tracker.init(livef, bbox) 

        def drawbox(livef, bbox):
            global cx, cy
            x, y, w, h= int(bbox[0]), (bbox[1]), (bbox[2]), (bbox[3])
            cv.rectangle(livef, (x, y), ((x+w), (y+h)), (255,0,255), 3, 1)
            cv.putText(livef, "Tracking", (75,75), cv.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)
            cx=x+w/2
            cy=y+h/2
            cv.circle(livef, (int(cx), int(cy)), radius=2, color=(255,0,255), thickness=-1)


        #ROIs selection:
        #ROIs = False
        start = False


        startPoint = False
        #endPoint = False
        assign = True


        #saving frames and finding points and calculations::
        label1 = True
        label2 = True
        count = 0
        calculations = False
        
        #hingepoint = False
        #checkpoint = True
        #Freqcal = True
        #freqframe = np.zeros(1)
        #frequency = np.ones(1)

        #detect local peaks:
        
        findmax = False
        findmin = True
        
        removeavemax0 = True
        
        removeavemin0 = True
        meanmax = False
        meanmin = False
        loopnum = 1
        loopmax = False
        loopmin = False


        CYELLOW  = '\033[33m'
        CRED = '\033[91m'
        CEND = '\033[0m'

        if emerexit2 == False:
            print("Press r to initiate the servo:")
            
        #size = (int(livevid.get(cv.CAP_PROP_FRAME_WIDTH)), int(livevid.get(cv.CAP_PROP_FRAME_HEIGHT)))
        
    while True:
        if emerexit == True:
            break
        if emerexit2 == True:
            break
        if emerexit3 == True:
            break
        if emerexit4 == True:
            break
            
        if keyboard.is_pressed("Esc"):
            emerexit5 = True
            finishind = True
            
        if emerexit5 == True:
            break
        
        liveret, livef = livevid.read()
        if liveret == True:

            #livef = cv.medianBlur(livef,5)    
            #livehsv = cv.cvtColor(livef, cv.COLOR_BGR2HSV)
            
            #To Zoom:
            if zoomed == True:
                height, width, channels = livef.shape

                minY = 368
                maxY = 768
                minX = 604
                maxX = 1315


                cropped = livef[minY:maxY , minX:maxX]
                livef = cv.resize(cropped, (width, height))
            
            
            success, bbox = tracker.update(livef)
            
            #TRACKING:
            
            if success:
                drawbox(livef, bbox)
                #print(cx, ", ", cy)
            else:
                cv.putText(livef, "lost", (75,75), cv.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
                print(CRED + "Tracking object is Lost; EMERGNCY EXIT ACTIVATED.", "" + CEND)
                emerexit = True
           
            cv.imshow("live video:", livef)
            #cv.imshow('live hsv:', livehsv)
            

                #Select ROIS:
            if start == False:
                if cv.waitKey(33) == ord('r'):
                    freqcalculations = True
                    start = True
                          


            if start == True and calculations == False:
                if label1 == True:
                    print("Now press c to do calculations.")
                    label1 = False
                #cv.rectangle(livef, (psp[0], psp[1]), (pep[0], pep[1]), (0, 0, 255), 2)
                if cv.waitKey(33) == ord('c'):
                    print("Calculations are being done.")
                    calculations = True
                    #print('im here1')
            if calculations == True:
                
                #save video:
                if recvid == True:
                    if label2 == True:
                        print("Video is being recorded")
                        label2 = False
                    savef.write(livef)

                ang = np.rad2deg(np.arctan((hx-cx)/(cy-hy)))

                if ang > 50:
                    print("extreme high angle is applied on the animal")
                    #ang = 45
                    emerexit4 = True
                if ang < -60:
                    print("extreme low angle is applied on the animal")
                    #ang = -55
                    emerexit4 = True
                #if keyboard.is_pressed("t"):
                 #   print("sythetic error generated")
                  #  emerexit4 = True
                
                
                
                angles = np.append(angles, ang)
                    
            
                #to find maximum angles:
                if findmax == True:
                    if angles[count] < angles[count-1]:
                        maxpeaks = np.append(maxpeaks, angles[count-1])
                        #print("\n relative max angle = ", maxpeaks[-1],"\n")
                        print(CRED + "detected max ang =", maxpeaks[-1],"" + CEND)
                        findmax =False
                        findmin = True
                        meanmax = True
                        
                        maxerror = True


                #to find minimum angles:
                if findmin == True:
                    if angles[count] > angles[count-1]:
                        minpeaks = np.append(minpeaks, angles[count-1])
                        #print("\n relative min angle = ", minpeaks[-1],"\n")
                        print(CRED + "detected min ang =", minpeaks[-1],"" + CEND)
                        findmin = False
                        findmax = True
                        meanmin = True
                        
                        minerror = True

                #to report the mean max points so far:
                if meanmax == True:
                    if maxpeaks[-1] > 5:
                        avemax = np.append(avemax, maxpeaks[-1])
                        if removeavemax0 == True:
                            avemax = np.delete(avemax, 0)
                            removeavemax0 = False
                        ave_max_ang = np.average(avemax)
                        print("average max:", ave_max_ang)
                        #print("Maximum of this loop:", maxpeaks[-1])
                        deltamax = MAX - maxpeaks[-1]
                        #print("delta max =", deltamax)
                        meanmax = False
                        loopmax = True
                    elif maxerror == True:
                        print(CRED + "too low max angle is detected", maxpeaks[-1],"" + CEND)
                        maxerror = False
                        deltamax = 0
                #to report the mean min points so far:
                if meanmin == True:
                    if minpeaks[-1] < -5:
                        avemin = np.append(avemin, minpeaks[-1])
                        if removeavemin0 == True:
                            avemin = np.delete(avemin, 0)
                            removeavemin0 = False
                        ave_min_ang = np.average(avemin)
                        print("average min:", ave_min_ang)
                        #print("MINIMUM of this loop:", minpeaks[-1])
                        deltamin = MIN - minpeaks[-1]
                        #print("delta min =", deltamin)
                        meanmin = False
                        loopmin = True
                    elif minerror == True:
                        print(CRED + "too low min angle is detected:", minpeaks[-1],"" + CEND)
                        minerror = False
                        deltamax = 0
                if loopmax == True and loopmin == True:
                    print(CYELLOW + "monitored Loop# %d:" % loopnum, "" + CEND)
                    loopnum = loopnum + 1
                    loopmax = False
                    loopmin = False
                count = count + 1
                
                    
                if cv.waitKey(33) == ord('q'):
                    finishind = True
                    #print('im here2')
                    break 

            pass

        else:
            print("Video is not availble anymore")
            break        


    livevid.release()
    print("Monitoring is stopped")
    #print("count=", count)
    if recvid == True:
        savef.release()
        #savem.release()
    cv.destroyAllWindows()
    
    #print('size:', size)

    if finishind == True:
        finish = True
        anglestotal = np.append(anglestotal, angles)
        anglestotal = np.append(anglestotal, angles)
        avemaxtotal = np.append(avemaxtotal, avemax) #effectives
        avemintotal = np.append(avemintotal, avemin) #effectives
        maxpeakstotal = np.append(maxpeakstotal, maxpeaks)
        minpeakstotal = np.append(minpeakstotal, minpeaks)
        
        print('monitoring data is saved')
        
        dfangles = pd.DataFrame(anglestotal)
        dfangles.rename(columns={0: 'angles'}, inplace=True)
        dfangles.to_excel("angles.xlsx", float_format = "%.3f", header = True)
        
        dfminpeaks = pd.DataFrame(minpeakstotal)
        dfminpeaks.rename(columns={0: 'minimums'}, inplace=True)
        dfminpeaks.to_excel("all min points.xlsx", float_format = "%.3f", header = True)
        
        dfmaxpeaks = pd.DataFrame(maxpeakstotal)
        dfmaxpeaks.rename(columns={0: 'maximums'}, inplace=True)
        dfmaxpeaks.to_excel("all max points.xlsx", float_format = "%.3f", header = True)
        
        dfavemin = pd.DataFrame(avemintotal)
        dfavemin.rename(columns={0: 'minimums'}, inplace=True)
        temp = np.zeros(avemintotal.size)
        dfavemin['averaged minimum'] = temp
        dfavemin['averaged minimum'][0] = avemintotal[0]
        for i in range (1,avemintotal.size):
            dfavemin['averaged minimum'][i] = (avemintotal[0:i+1].sum())/(i+1)
        dfavemin.to_excel("effective min points.xlsx", float_format = "%.3f", header = True)
        
        dfavemax = pd.DataFrame(avemaxtotal)
        dfavemax.rename(columns={0: 'maximums'}, inplace=True)
        temp = np.zeros(avemaxtotal.size)
        dfavemax['averaged maximums'] = temp
        dfavemax['averaged maximums'][0] = avemaxtotal[0]
        for i in range (1,avemaxtotal.size):
            dfavemax['averaged maximums'][i] = (avemaxtotal[0:i+1].sum())/(i+1)
        dfavemax.to_excel("effective max points.xlsx", float_format = "%.3f", header = True)
    
    if finishind == False:
        anglestotal = np.append(anglestotal, angles)
        anglestotal = np.append(anglestotal, angles)
        avemaxtotal = np.append(avemaxtotal, avemax) #effectives
        avemintotal = np.append(avemintotal, avemin) #effectives
        maxpeakstotal = np.append(maxpeakstotal, maxpeaks)
        minpeakstotal = np.append(minpeakstotal, minpeaks)
    
    repeat = True
    
```
